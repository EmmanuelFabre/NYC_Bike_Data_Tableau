```python
import pandas as pd 
```

```python
sumdf = pd.read_csv("201808-citibike-tripdata.csv")
sumdf.head()
```

```python
windf = pd.read_csv("201802-citibike-tripdata.csv")
windf.head()
```

```python
combdf = pd.merge(sumdf, windf, on= "starttime")
combdf.head()
```

```python
summer = sumdf.drop(columns=['tripduration', 'birth year', 'gender', 'bikeid'])
summer.head()



```

```python
winter = windf.drop(columns=['tripduration', 'birth year', 'gender', 'bikeid'])
winter.head()
```

```python
comb = pd.concat([summer, winter])
comb.head()
```

```python
comb.to_csv("workingfile.csv")
```
